```python
def velocity_dispersion(m_gal_kg):
    """Calculate velocity dispersion in km/s using scaling relation"""
    m_gal_ratio = m_gal_kg / (1e11 * const.M_sun.to(u.kg).value)
    v_kms = (10**2.2969) * m_gal_ratio**0.299  # km/s
    return v_kms  # now returns float [km/s]

def coulomb_logarithm(r_km, sigma_kms, G_km3_kg_s2, m_kg):
    """Coulomb logarithm in km units"""
    x = (r_km * sigma_kms**2) / (G_km3_kg_s2 * m_kg)
    return math.log10(x.value)

def dynamical_friction_acceleration(r_mag_km, v_vec_kms, ln_Lambda, G_, m_kg, sigma_kms):
    """Compute DF acceleration in km/s²"""
    v_mag = np.linalg.norm(v_vec_kms)
    if v_mag == 0:
        return np.zeros(3)

    unit_v = v_vec_kms / v_mag

    # Chandrasekhar DF magnitude, km/s²
    a_df_mag = (4 * np.pi * G_**2 * m_kg * ln_Lambda * sigma_kms**2) / (v_mag**3)
    return -a_df_mag * unit_v



def calculate_DF(G_val, m1_val, m2_val, m_gal_val, r1_vec_km, r2_vec_km, v1_vec_kms, v2_vec_kms):
    """Compute DF for both BHs"""
    sigma = velocity_dispersion(m_gal_val * u.kg)
    
    r1_mag = np.linalg.norm(r1_vec_km)
    r2_mag = np.linalg.norm(r2_vec_km)
    
    ln_L1 = coulomb_logarithm(r1_mag, sigma, G_val, m1_val)
    ln_L2 = coulomb_logarithm(r2_mag, sigma, G_val, m2_val)
    
    a1_df = dynamical_friction_acceleration(r1_mag, v1_vec_kms, ln_L1, G_val, m1_val, sigma)
    a2_df = dynamical_friction_acceleration(r2_mag, v2_vec_kms, ln_L2, G_val, m2_val, sigma)
    
    return a1_df, a2_df
# def sigma(m_gal_):
#     """
#     Calculate the velocity dispersion of the galaxy.
#     Output: km/s (astropy Quantity)
#     """
#     return (10**2.2969 * (m_gal_ / (1e11 * const.M_sun))**0.299) * (u.km / u.s)

# def coulomb_log(dist_, o_, G_, m_):
#     """
#     Coulomb logarithm using physical units.
#     dist_: distance from center [length unit]
#     o_: velocity dispersion [speed unit]
#     G_: gravitational constant [length^3 / (mass * time^2)]
#     m_: mass [mass]
#     """
#     x = (dist_ * o_**2) / (G_ * m_)
#     return math.log10(x.to_value(u.dimensionless_unscaled))

# def DF_force(ln_Lambda, m_, m_gal_, r_choice_, rcom_, vfirst_, vsecond_, G_):
#     """
#     Chandrasekhar Dynamical Friction Force (returns vector in kg·m/s²).
#     """
#     r_rel = r_choice_ - rcom_
#     v_rel = vfirst_ - vsecond_
#     v_mag = np.linalg.norm(v_rel)

#     if v_mag == 0:
#         return np.zeros(3) * u.kg * u.m / u.s**2

#     v_hat = v_rel / v_mag
#     r_mag = np.linalg.norm(r_rel)

#     F_mag = 0.428 * ln_Lambda * (G_ * m_**2 / r_mag**2)
#     F_vec = -F_mag * v_hat

#     print(F_vec)

#     return F_vec 

# def twobodyeqn(t, w, G_, m1_, m2_, m_gal_, rcom_, r_, df_on=False):
#     """
#     Two-body equation with dynamical friction.
#     Inputs:
#       - G_: gravitational constant [km^3 / kg / s^2]
#       - m1_, m2_: masses [kg]
#       - w: state vector [r1, r2, v1, v2]
#     Outputs:
#       - dy/dt array with positions and velocities
#     """
#     # Unpack state vector
#     r1_ = w[0:3] * u.km
#     r2_ = w[3:6] * u.km
#     v1_ = w[6:9] * u.km / u.s
#     v2_ = w[9:12] * u.km / u.s

#     r_vec = r2_ - r1_
#     r = np.linalg.norm(r_vec)

#     o = sigma(m_gal_)

#     dist1 = np.linalg.norm(r1_ - rcom_)
#     dist2 = np.linalg.norm(r2_ - rcom_)

#     ln_A1 = coulomb_log(dist1, o, G_, m1_)
#     ln_A2 = coulomb_log(dist2, o, G_, m2_)


#     df_force1 = DF_force(ln_A1, m1_, m_gal_, r1_, rcom_, v2_, v1_, G_)
#     df_force2 = DF_force(ln_A2, m2_, m_gal_, r2_, rcom_, v1_, v2_, G_)

#     df_accel1 = (df_force1 / m1_).to(u.km / u.s**2)
#     df_accel2 = (df_force2 / m2_).to(u.km / u.s**2)


#     grav_accel1 = (G_ * m2_ * r_vec / r**3).to(u.km / u.s**2)
#     grav_accel2 = (-G_ * m1_ * r_vec / r**3).to(u.km / u.s**2)


#     dv1bydt = grav_accel1 + df_accel1
#     dv2bydt = grav_accel2 + df_accel2


#     dr1bydt = v1_.to(u.km / u.s)
#     dr2bydt = v2_.to(u.km / u.s)


#     return np.concatenate([
#         dr1bydt.to_value(u.km / u.s),
#         dr2bydt.to_value(u.km / u.s),
#         dv1bydt.to_value(u.km / u.s**2),
#         dv2bydt.to_value(u.km / u.s**2)
#     ])
```
